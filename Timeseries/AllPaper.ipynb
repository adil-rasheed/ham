```python
# The function generates a sinusoidally varying temperature profile for a 24 hours period
def generate_temperature_profile():
    import numpy as np
    import matplotlib.pyplot as plt

    # Define the parameters for the sinusoidal function
    hours = np.arange(0, 24, 0.1)  # Time in hours
    amplitude = 2 # Amplitude of the temperature variation
    mean_temperature = 20  # Mean temperature in degrees Celsius
    period = 24  # Period of the sinusoidal function (24 hours)

    # Generate the temperature profile using a sinusoidal function
    temperature_profile = mean_temperature + amplitude * np.sin((2 * np.pi / period) * (hours - 6))

    # Plot the temperature profile
    plt.figure(figsize=(10, 3))
    plt.plot(hours, temperature_profile, label='Temperature Profile', color='blue')
    plt.title('24-Hour Temperature Profile')
    plt.xlabel('Time (hours)')
    plt.ylabel('Temperature (Â°C)')
    plt.grid()
    plt.legend()
    plt.show()
    return hours, temperature_profile
# Call the function to generate and plot the temperature profile
generate_temperature_profile()
```

```python
#superimpose some oscillation to the temperature profile and replot
def generate_temperature_profile_with_oscillation():
    import numpy as np
    import matplotlib.pyplot as plt

    # Define the parameters for the sinusoidal function
    hours = np.arange(0, 24, 0.1)  # Time in hours
    amplitude = 2 # Amplitude of the temperature variation
    mean_temperature = 20  # Mean temperature in degrees Celsius
    period = 24  # Period of the sinusoidal function (24 hours)

    # Generate the temperature profile using a sinusoidal function
    temperature_profile = mean_temperature + amplitude * np.sin((2 * np.pi / period) * (hours - 6))

    # Add some oscillation to the temperature profile
    oscillation_amplitude = 0.5
    oscillation_frequency = 0.5  # Frequency of the oscillation
    oscillation = oscillation_amplitude * np.sin((2 * np.pi * oscillation_frequency) * hours)

    # Superimpose the oscillation on the temperature profile
    temperature_profile_with_oscillation = temperature_profile + oscillation

    # Plot the temperature profile with oscillation
    plt.figure(figsize=(10, 3))
    plt.plot(hours, temperature_profile_with_oscillation, label='Temperature Profile with Oscillation', color='red')
    plt.title('24-Hour Temperature Profile with Oscillation')
    plt.xlabel('Time (hours)')
    plt.ylabel('Temperature (Â°C)')
    plt.grid()
    plt.legend()
    plt.show()
    return hours, temperature_profile_with_oscillation
# Call the function to generate and plot the temperature profile with oscillation
generate_temperature_profile_with_oscillation()
```

```python
# Re-import necessary libraries after code execution state reset
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate the temperature profile with oscillation and noise
def generate_temperature_data():
    hours = np.arange(0, 24, 0.1)
    amplitude = 2
    mean_temperature = 20
    period = 24
    temperature_profile = mean_temperature + amplitude * np.sin((2 * np.pi / period) * (hours - 6))
    
    oscillation_amplitude = 0.5
    oscillation_frequency = 0.5
    oscillation = oscillation_amplitude * np.sin((2 * np.pi * oscillation_frequency) * hours)
    
    noise_amplitude = 0.5
    noise = noise_amplitude * np.random.normal(size=hours.shape)
    
    data = temperature_profile + oscillation + noise
    plt.figure(figsize=(10, 3))
    plt.plot(hours, data, label='Temperature Profile with Oscillation and Noise', color='purple')
    plt.title('24-Hour Temperature Profile with Oscillation and Noise')
    plt.xlabel('Time (hours)')
    plt.ylabel('Temperature (Â°C)')
    plt.grid()
    plt.legend()
    plt.show()
    return hours, data

# Apply detrended fluctuation analysis (DFA)
def detrended_fluctuation_analysis(data, min_box_size=4, max_box_size=50):
    n = len(data)
    integrated_data = np.cumsum(data - np.mean(data))

    box_sizes = np.floor(np.logspace(np.log10(min_box_size), np.log10(max_box_size), num=20)).astype(int)
    flucts = []

    for box_size in box_sizes:
        n_boxes = n // box_size
        rms = []

        for i in range(n_boxes):
            segment = integrated_data[i * box_size:(i + 1) * box_size]
            x = np.arange(box_size)
            coeffs = np.polyfit(x, segment, 1)
            trend = np.polyval(coeffs, x)
            rms.append(np.sqrt(np.mean((segment - trend) ** 2)))
        
        flucts.append(np.mean(rms))

    flucts = np.array(flucts)
    coeffs = np.polyfit(np.log(box_sizes), np.log(flucts), 1)
    alpha = coeffs[0]

    return box_sizes, flucts, alpha

# Run DFA on the temperature data
hours, temperature_data = generate_temperature_data()
box_sizes, flucts, alpha = detrended_fluctuation_analysis(temperature_data)

# Plot the DFA result
plt.figure(figsize=(6, 4))
plt.plot(np.log(box_sizes), np.log(flucts), 'o-', label=f'Î± â {alpha:.2f}')
plt.xlabel('log(Box size)')
plt.ylabel('log(Fluctuation)')
plt.title('Detrended Fluctuation Analysis (DFA)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from numpy.linalg import svd

# -----------------------
# MSSA Helper Functions
# -----------------------
def embed_time_series(X, M):
    L, N = X.shape
    K = N - M + 1
    trajectory = np.zeros((L * M, K))
    for l in range(L):
        for i in range(M):
            trajectory[l * M + i] = X[l, i:i + K]
    return trajectory

def compute_dfa(signal, min_scale=4, max_scale=50):
    N = len(signal)
    scales = np.floor(np.logspace(np.log10(min_scale), np.log10(max_scale), num=20)).astype(int)
    fluctuations = []
    profile = np.cumsum(signal - np.mean(signal))

    for scale in scales:
        segments = N // scale
        rms = []
        for i in range(segments):
            idx = slice(i * scale, (i + 1) * scale)
            segment = profile[idx]
            x = np.arange(scale)
            coeffs = np.polyfit(x, segment, deg=1)
            trend = np.polyval(coeffs, x)
            rms.append(np.sqrt(np.mean((segment - trend) ** 2)))
        fluctuations.append(np.mean(rms))

    log_scales = np.log(scales)
    log_flucts = np.log(fluctuations)
    slope, _ = np.polyfit(log_scales, log_flucts, 1)
    return slope

# -----------------------
# MSSA-DFA Denoising Core
# -----------------------
def mssa_dfa_denoise(X, M=30, dfa_threshold=1.0):
    L, N = X.shape
    trajectory = embed_time_series(X, M)
    U, Sigma, VT = svd(trajectory, full_matrices=False)

    RCs = []
    for k in range(U.shape[1]):
        Uk = U[:, k].reshape((L, M))
        Sk = Sigma[k]
        Vk = VT[k, :]
        RC_k = np.zeros((L, N))
        for l in range(L):
            for i in range(N - M + 1):
                RC_k[l, i:i + M] += Sk * Uk[l, :] * Vk[i]
        RC_k /= M
        RCs.append(RC_k)

    # DFA filtering
    selected = []
    for k, RC_k in enumerate(RCs):
        keep = False
        for l in range(L):
            alpha = compute_dfa(RC_k[l])
            if alpha > dfa_threshold:
                keep = True
        if keep:
            selected.append(RC_k)

    if not selected:
        print("â ï¸ No components passed DFA threshold. Try lowering the threshold.")
        return np.zeros_like(X)

    denoised = np.sum(selected, axis=0)
    return denoised
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_temperature_profile_with_realistic_stiction():
    hours = np.arange(0, 24, 0.1)
    amplitude = 2
    mean_temperature = 20
    period = 24

    # Ideal sinusoidal temperature
    temperature_profile = mean_temperature + amplitude * np.sin((2 * np.pi / period) * (hours - 6))

    # Stiction model
    stiction_threshold = 0.1  # Minimum control effort to move
    stiction_profile = np.zeros_like(temperature_profile)
    last_position = temperature_profile[0]
    control_effort = 0  # Accumulates control input until threshold is exceeded

    for i in range(1, len(temperature_profile)):
        error = temperature_profile[i] - last_position
        control_effort += error

        if abs(control_effort) >= stiction_threshold:
            last_position += control_effort
            control_effort = 0  # Reset after breaking free

        stiction_profile[i] = last_position

    # Plot
    plt.figure(figsize=(12, 4))
    plt.plot(hours, temperature_profile, '--', label='Ideal Profile', color='blue')
    plt.plot(hours, stiction_profile, label='With Realistic Stiction', color='red')
    plt.title('Temperature Profile with Realistic Stiction Effect')
    plt.xlabel('Time (hours)')
    plt.ylabel('Temperature (Â°C)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return hours, temperature_profile, stiction_profile

```
